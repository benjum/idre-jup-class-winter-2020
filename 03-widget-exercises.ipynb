### Reverse some text

Here is a function that takes text as an input and returns the text backwards.

```python
def reverse(x):
    return x[::-1]

reverse('I am printed backwards.')
```

### Make a plot

Here is a python function that, given $k$ and $p$, plots $f(x) = \sin(k x + p)$.


```python
def plot_f(k, p):
    x = np.linspace(0, 4 * np.pi)
    y = np.sin(k*x + p)
    plt.plot(x, y)
    
plot_f(1.3, 3)
```
